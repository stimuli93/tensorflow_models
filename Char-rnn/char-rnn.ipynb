## Character Modelling Using Shakespeare text for training

```python
import numpy as np
```

### Loading text from all of the Shakespeare's text

```python
chars = []
for line in open('dataset/shakespeare.txt'):
    chars += list(line)
```

```python
chars
```

```python
corpus_size = len(chars)
unique_chars = set(chars)
vocab_size = len(unique_chars)
print 'Corpus Size: %s , Vocab Size: %s' % (corpus_size, vocab_size)
```

```python
char_to_index ={w:i for i,w in enumerate(unique_chars)}
index_to_chars = {i:w for i,w in enumerate(unique_chars)}
```

```python
trX = [char_to_index[character] for character in chars]
```

```python
from char_rnn import CharRNN
rnn = CharRNN(max_seq_length=128,vocab_size=vocab_size, batch_size=16)
```

```python
rnn.train(trX, n_iters=2000,learning_rate=2e-3)
```

```python
text = rnn.sample(trX[0:25],char_count=200)
```

```python
chars = [index_to_chars[i] for i in text]
```

### Generating a 200 word document from the learnt parameters

```python
print(''.join(chars))
```
